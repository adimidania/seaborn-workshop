```python
import seaborn as sns 
import numpy as np
import pandas as pd

sns.set_style('darkgrid')
```

```python
tips_data = sns.load_dataset("tips")
```

```python
tips_data.info()
```

```python
tips_data.head()
```

# Categorical plots:

## Count plot:

```python
# Males and Females
sns.catplot(kind='count', data=tips_data, x="sex")
```

```python
# Smokers and Non Smokers
sns.catplot(kind='count', data=tips_data, x="smoker")
```

## Bar plot:

```python
sns.catplot(kind='bar', data=tips_data, x="sex", y="total_bill")
```

## Strip plot:

```python
sns.catplot(kind='strip', data=tips_data, x="sex", y="total_bill", jitter=0.2)
```

## Swarm plot:

```python
sns.catplot(kind='swarm', data=tips_data, x="sex", y="total_bill")
```

## Box plot:

```python
sns.catplot(kind='box', data=tips_data, x="time", y="total_bill")
```

## Violin plot:

```python
# Each stick represents an observation
# We can use quartiles in the inner stick so we can represent our quartiles
sns.catplot(kind='violin', data=tips_data, x="time", y="total_bill", inner="stick")
```

# Distibution plots:

## Hist plot:

```python
# Variables are continuous
sns.displot(kind='hist', data=tips_data, x="total_bill", bins=30)
```

## Kde plot:

```python
# Using the kernel density estimation
sns.displot(kind='kde', data=tips_data, x="total_bill")
```

## Kde plot with two variables:

```python
sns.displot(kind='kde', data=tips_data, x="total_bill", y="tip")
```

## Ecdf plot:

```python
sns.displot(kind='ecdf', data=tips_data, x="total_bill")
```

# Relational plot:

## Scatter plot:

```python
# The variables are continuous
sns.relplot(kind='scatter', data=tips_data, x="total_bill", y="tip")
```

## Line plot:

```python
sns.relplot(kind='scatter', data=tips_data, x="total_bill", y="tip")
```
